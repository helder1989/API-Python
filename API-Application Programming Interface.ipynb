## O que seria uma API? 

 APIs são mecanismos que permitem que dois componentes de software se comuniquem usando um conjunto de definições e protocolos. Muitas empresas, utilizam as APIs para facilitar a obtenção ou troca de informações. Quando vamos realizar login ou nos cadastrar em algum site, a página de acesso pode utilizar-se das nossas contas já salvas no servidor para obter nossos dados e assim para concluir o login/cadastro no site, por exemplo. Quando isso acontece é uma API disponibilizando opção de logar com sua conta do Facebook ou e-mail já existentes ali.
 
No exemplo acima, destacamos que existem diversos sites que nos permitem usar o Facebook como forma de login sem ter que realizar cadastro algum. O que o Facebook fez foi um programa para que esse site possa obter algumas de nossas informações para nos cadastrar sem que tenha que digita-lás. Cabe ressaltar que o site não pode obter tudo que ele quiser em relação aos dados de uma conta. O mesmo só vai ter acesso ao que o Facebook disponibilizar dentro do código de programação dessa API, então eles vão definir os dados que serão obtidos ou trocados.

Nesse estudo vamos verificar como usar API em Python para que possamos automatizar algumas atividades dentro do nosso código, no caso será uma API (Pública e gratuita) de cotações de moedas.

## Quais são os benefícios gerais das APIs? 

As APIs apresentam uma oportunidade única para as empresas atenderem às necessidades de seus clientes em diferentes plataformas. Um exemplo é a API do Linkedin, que permite a integração de informações de vagas por meio de sites. Qualquer empresa pode fornecer acesso semelhante aos seus respectivos bancos de dados internos usando APIs gratuitas ou pagas.

## Quais são os diferentes tipos de API? 

As APIs são classficadas de acordo com sua arquitetura e escopo de uso.

## APIs privadas

Elas são internos a uma empresa e são usadas apenas para conectar sistemas e dados dentro da empresa;

## APIs públicas 

Estas são abertas ao público e podem ser usadas por qualquer pessoa. Pode ou não haver alguma autorização e custo associado a esses tipos de APIs;

## APIs de parceiros 

Estas são acessíveis apenas por desenvolvedores externos autorizados para auxiliar as parcerias entre empresas;

## APIs compostas 

Estas combinam duas ou mais APIs distintas para atender a requisitos ou comportamentos complexos do sistema.

## Mão na massa: Momento no qual vou extrair a cotação das seguintes moedas: Dólar, Euro e Bitcoin.

Link usado: awesome api: https://docs.awesomeapi.com.br/

## Dólar :

```
# Fazer requisição
import requests 
import json 

cotacoes = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")
cotacoes = cotacoes.json()#transformando de formato json para o formato python padrão
cotacao_dolar = cotacoes['USDBRL']["bid"]
#print(cotacoes)
print(cotacao_dolar)
```

## Euro :

```
cotacoes = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")
cotacoes = cotacoes.json()#transformando de formato json para o formato python padrão
cotacao_euro = cotacoes['EURBRL']["bid"]
#print(cotacoes)
print(cotacao_euro)
```

## Bitcoin :

```
cotacoes = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")
cotacoes = cotacoes.json() #transformando de formato json para o formato python padrão
cotacao_bitcoin = cotacoes['BTCBRL']["bid"]
#print(cotacoes)
print(cotacao_bitcoin)
```
